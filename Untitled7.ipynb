```python
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt

f = open('sortData.txt', 'r')
results = []
for i in f:
        k = len(i)
        results.append(k)
results.sort()
print(results)



# NOTE: for *great* examples of all things matplotlib-related, see:
# http://matplotlib.sourceforge.net/gallery.html

x_positions = range(len(results))
plt.bar(x_positions, results, align='center', alpha=0.5)
plt.xticks(x_positions, i)
plt.xlabel('letter')
plt.ylabel('Number')
plt.title('Number of words that start with a given letter')
plt.show()
```
